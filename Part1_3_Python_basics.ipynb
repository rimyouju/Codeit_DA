## (2) 자료형

**자료형(Data Type)**
- 숫자(Number) : 수학적 계산이 가능한 데이터
  - 정수(Integer) : 소수점을 포함하지 않는 숫자형 데이터 `예) 10, -5, 0`
  - 실수(Floating Point) : 소수점을 포함하는 숫자형 데이터 `예) 3.14, -0.001, 2.0`

## (4) 제어문

### (a) while

while문은 무언가를 반복하기 위해 사용한다.

**while 반복문 구조**

```Python
while 조건 부분:  # 불린 값으로 계산되는 식
    수행 부분     # 반복적으로 실행하고 싶은 명령
```

````python
i = 1

while i <= 3:
    print(i)
    i += 1  # i = i + 1
````

````python
i = 1

while i <= 3:
    print("나는 잘 생겼다!")
    i += 1
````

## (10) 예외 처리

### try-except

- `try-except` : 프로그램의 흐름을 안전하게 유지하는 방법으로 `try-except` 예외 처리 방법을 적용할 수 있다.

```
try:
    # 에러가 발생할 수 있는 코드
except 에러종류:
    # 에러가 발생했을 때 실행할 코드
```

- `try` : 일단 시도해본다
- `except` : 실패하면 이렇게 처리한다는 의미로, 에러가 없으면 except는 실행되지 않는다.

````python
x = int(input("숫자를 입력하세요: "))  # 문자를 입력하면, ValueError
````

````python
try:
    x = int(input("숫자를 입력하세요: "))
    print(x * 2)
except ValueError:
    print("숫자가 아닙니다. 숫자를 입력하세요.")
````

## for문 퀴즈

---
실습) for문으로 구구단

[링크](https://www.codeit.kr/topics/python-programming-and-data/lessons/2801)

````python
for i in range(1, 10):
    for j in range(1, 10):
        print("{} * {} = {}".format(i, j, i*j))
````

````python
# while 문 사용
i = 1
while i < 10:
    j = 1
    while j < 10:
        print("{} * {} = {}".format(i, j, i * j))
        j += 1
    i += 1
````
