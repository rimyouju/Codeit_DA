## (1) array

### array 생성하기

- `np.array()` : 파이썬의 리스트(list)나 튜플(tuple) 등을 numpy의 ndarray 객체로 변환하는 함수
- `np.zeros(n)` : 0이 n개 들어있는 배열 생성  
- `np.ones(n)` : 1이 n개 들어있는 배열 생성  
- `np.arange(a)` : 0 이상 a 미만의 정수 배열 생성  
- `np.arange(a, b)` : a 이상 b 미만의 정수 배열 생성  
- `np.arange(a, b, c)` : a 이상 b 미만, c 간격의 배열 생성

````python
# 라이브러리 불러오기
import numpy as np
````

- `np.random.randint`(a, b, size=n) : a 이상 b 미만 n개 정수 난수 생성
- `np.random.choice`(list, size=n) : 리스트에서 n개 샘플 추출
- `np.random.normal`(loc, scale, size=n) : 평균(loc), 표준편차(scale)의 정규 분포 난수 n개 생성
- `np.linspace`(a, b, num=n) : a부터 b까지 n개 균등 분할, 일정 간격의 수열 생성

````python
# 0 이상 10 미만 정수 5개
np.random.randint(0, 10, size=5)
````

````python
# 리스트에서 3개 랜덤 추출(기본값 : replace=True(중복 허용 추출))
np.random.choice(['A', 'B', 'C', 'D'], size=3)
````

## (2) 조건에 맞는 데이터 선택하기

- 불린 인덱싱(`Boolean Indexing`) : 조건식의 결과로 나온 `True/False`값을 이용해, 원하는 데이터만 선택하는 방법
- 원리: 조건을 만족하는 행(또는 열)은 `True`로 표시하고, 조건을 만족하지 않으면 `False`로 숨김
- 예시: 특정 값 이상인 데이터, 특정 조건을 만족하는 행만 필터링

(예시1)

````python
gdp_array = np.array([6610, 7637, 8127, 8885, 10385, 12565, 14403, 12398, 8282, 10672])
gdp_array
````

````python
# gdp가 10000 초과인 경우
gdp_array[gdp_array > 10000]
````

````python
mask = gdp_array > 10000
gdp_array[mask]
````

(예시2)

````python
gdp_array
````

````python
# gdp가 8000 이상이고, 10000 이하인 경우
gdp_array[(gdp_array >= 8000) & (gdp_array <= 10000)]
````

````python
condition = (gdp_array >= 8000) & (gdp_array <= 10000)
gdp_array[condition]
````

````python
condition1 = (gdp_array >= 8000)
condition2 = (gdp_array <= 10000)
gdp_array[condition1 & condition2]
````

(예시3)

````python
gdp_array
````

````python
# gdp가 8000 미만 이거나 gdp가 10000 초과인 경우
gdp_array[(gdp_array < 8000) | (gdp_array > 10000)]
````

````python
condition = (gdp_array < 8000) | (gdp_array > 10000)
gdp_array[condition]
````

````python
condition1 = (gdp_array < 8000)
condition2 = (gdp_array > 10000)
gdp_array[condition1 | condition2]
````

| 옵션           | 설명                                                                 | 기본값 |
|----------------|----------------------------------------------------------------------|--------|
| subset         | 중복 여부를 확인할 **특정 컬럼** 지정 가능 (리스트 형태도 가능)        | None (전체 컬럼 기준) |

| ㆍ 'first'     | 처음 등장한 행은 남기고, 이후 중복된 행을 True로 표시                 |        |
| ㆍ 'last'      | 마지막 등장한 행은 남기고, 이전 중복된 행을 True로 표시               |        |
| ㆍ False       | 중복된 모든 행을 True로 표시                                          |        |
| inplace        | DataFrame에 바로 적용할지 여부 (실제 `drop_duplicates()`에서 주로 사용) | False  |

### (b) 스케일링 : 정규화, 표준화

**스케일링(Scaling)**  
   - 숫자 데이터의 **단위나 범위를 일정하게 맞추는 과정**으로 서로 다른 단위를 가진 데이터를 비교 가능하게 만든다.

(정규화)

**정규화(Normalization)**  
   - 데이터를 **0과 1 사이 값**으로 변환하는 것으로 **Min-Max Normalization**이라고도 부른다.

**정규화 계산식**
   - 데이터의 최솟값은 0, 최댓값은 1로 맞추고 나머지 값들은 그 사이 비율에 맞게 변환한다.
   - 계산식:  
   $$
   \text{정규화} = \frac{(x - \min(x))}{\max(x) - \min(x)}
   $$

````python
patient_df.head()
````

````python
# height 컬럼 정규화
patient_df['height'] = (patient_df['height'] - patient_df['height'].min()) / (patient_df['height'].max() - patient_df['height'].min())
````

````python
patient_df.head()
````

````python
# 정규화 확인 : 최솟값과 최댓값이 각각 0과 1
patient_df['height'].describe()
````

(표준화)

**표준화(Standardization, Standard Scaling)**  
   - 각 값이 **평균에 비해 얼마나 큰지 또는 작은지**를 나타내는 것으로 데이터를 변환했을 때 평균은 0, 분산은 1이 되도록 맞춘다.

**z-score 계산식**

$$
z = \frac{(x - \mu)}{\sigma}
$$

   - 데이터가 평균에서 얼마나 떨어져 있는지(거리)를 표준편차 단위로 표현한다.

````python
# 데이터 불러오기
patient_df = pd.read_csv('patient.csv')
patient_df.head()
````

````python
# height 컬럼 표준화
patient_df['height'] = (patient_df['height'] - patient_df['height'].mean()) / (patient_df['height'].std())
````

````python
# 표준화 확인 : mean(평균), std(표준편차) 각각 0과 1
patient_df['height'].describe()
````

**데이터 스케일링 정리**

1. **정규화 (Normalization)**  
  - 모든 값을 0과 1 사이 값으로 변환하여 데이터의 최솟값은 0, 최댓값은 1이 되도록 조정  

2. **표준화 (Standardization)**  
  - 각 데이터가 평균에서 얼마나 떨어져 있는지를 비교하여 평균 = 0, 표준편차 = 1이 되도록 조정

### (b) 다양한 통계 계산

- `groupby`는 그룹별로 다양한 집계 함수(`size`, `count`, `mean`, `sum`, `max`, `min` 등)를 적용할 수 있으며, `agg()`를 사용하면 여러 통계값을 한 번에 계산할 수 있다.
  - `size` : 전체 행 개수
  - `count` : 값 개수
  - `mean`, `sum`, `max`, `min` : 대표적인 수치 집계 함수
  - `agg` : 여러 통계 지표 동시에 계산 가능

````python
netflix_df.head()
````

````python
# 장르별 평점
netflix_df.groupby(['genre'])['score'].mean()
````

````python
# 장르별 평점(agg 사용)
netflix_df.groupby(['genre'])['score'].agg('mean')
````

````python
# 장르별 점수의 평균, 최댓값, 최솟값(agg 사용)
netflix_df.groupby(['genre'])['score'].agg(['mean', 'max', 'min'])
````

````python
# 장르별 점수의 평균, 런타임의 최솟값과 최댓값
netflix_df.groupby(['genre']).agg({'score':'mean', 'runtime':['min', 'max']})
````

  - `by`: 정렬할 컬럼 지정
  - `ascending=True`: 오름차순(기본값)
  - `ascending=False`: 내림차순

````python
# 장르별 점수의 평균, 런타임의 평균
netflix_df.groupby(['genre']).agg({'score':'mean', 'runtime':'mean'}).sort_values(by='score', ascending=False)
````

````python
pd.pivot_table(
    olympic_df,
    index='team',
    columns='medal',
    aggfunc='size'
).fillna(0).sort_values(
    by=['1st', '2nd', '3rd', 'team'],
    ascending=False)
````
