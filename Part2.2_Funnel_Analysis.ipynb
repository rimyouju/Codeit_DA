# 퍼널 분석용 데이터 준비

- 퍼널 분석(Funnel Analysis) : 고객이 특정 목표에 도달하기까지 거치는 일련의 단계를 분석하는 기법
- 퍼널 분석을 활용하면 특정 목표에 도달하기까지 어떤 단계에 개선이 필요한지 확인할 수 있어 웹 사이트나 앱 내의 화면 기획 시 주요 참고 자료로 활용 가능

```python
cosmetic_df.head()
```

```python
# user_id, product_id, event_time을 기준으로 데이터 정렬
cosmetic_df = cosmetic_df.sort_values(by=['user_id', 'product_id', 'event_time'])
```

```python
# user_id, product_id 기준으로 첫 이벤트
temp = cosmetic_df.groupby(['user_id', 'product_id']).first()
```

```python
temp.head()
```

```python
# user_id - product_id 조합의 첫 이벤트가 view인 경우만
target_index = temp.query("event_type == 'view'").index
```

```python
target_index
```

```python
# user_id - product-id 조합의 첫 이벤트가 view인 경우
cosmetic_df = cosmetic_df.set_index(['user_id', 'product_id']).loc[target_index].reset_index()
```

# 퍼널 분석

- 목표 : 고객의 구매(purchase) 촉진
- 퍼널 단계 : view -> cart -> purchase

## 단계별 전환율

```python
cosmetic_df.head()
```

```python
# 각 단계별 unique한 product_id - user_id 조합 집계
user_counts = cosmetic_df.groupby(['event_type', 'product_id'])['user_id'].nunique()
```

```python
user_counts
```

```python
# 각 단계별 유저 수 집계
user_counts = user_counts.groupby('event_type').sum().reset_index()
```

```python
# 컬럼명 변경(user_id -> users)
user_counts = user_counts.rename(columns={'user_id' : 'users'})
```

```python
# 'remove_from_cart' 단계 제외
user_counts = user_counts.query("event_type != 'remove_from_cart'")
```

```python
# 퍼널 단계 순으로 정렬(유저 수가 많은 단계에서 적은 단계 순으로 정렬)
user_counts = user_counts.sort_values(by='users', ascending=False)
```

```python
user_counts
```

```python
# 퍼널 시각화
px.funnel(user_counts,
          x='users',
          y='event_type')
```

```python
user_counts
```

```python
# 전환율 계산
user_counts['conversion_rate(%)'] = user_counts['users'] / user_counts['users'].shift(1) * 100
```

```python
# 이탈률 계산
user_counts['churn_rate(%)'] = 100 - user_counts['conversion_rate(%)']
```

```python
user_counts
```

- 유저 관점에서 상품 탐색의 흐름을 봤을 때, view에서 cart로 넘어가는 단계에서 약 87.5%가 이탈하고, cart에서 purchase로 넘어가는 단계에서는 약 63.5%가 이탈하고 있다.

## view 100% 기준 전환율

```python
user_counts
```

```python
# view 100%
user_counts.iloc[0, 1]
```

```python
# view를 100%로 두고 전환율 계산
user_counts['conversion_rate_total'] = user_counts['users'] / user_counts.iloc[0, 1]
```

```python
user_counts
```

```python
# 퍼널 시각화
fig = px.funnel(user_counts,
                x='conversion_rate_total',
                y='event_type')
fig.update_traces(texttemplate="%{value:.2%}")
```

- 상품을 본 후, 장바구니에 담기까지의 전환율은 약 12.5%이고,
- 상품을 본 후, 구매까지 이어지는 전환율은 약 4.6%임을 확인할 수 있다.

## 가격대별 분석

- 상품의 가격대에 따라 전환율에 차이가 있는지 확인하고자 한다.

```python
cosmetic_df.head()
```

```python
# 상품별 가격 데이터 정리
temp = cosmetic_df[['product_id', 'price']].drop_duplicates()
```

```python
temp['price'].quantile([0.25, 0.5, 0.75])
```

```python
# 가격대 함수
def group_price(x):
    if x <= 3:
        return '~3'
    elif x <= 5:
        return '3~5'
    elif x <= 8:
        return '5~8'
    else:
        return '8~'
```

```python
cosmetic_df['price_range'] = cosmetic_df['price'].apply(group_price)
```

```python
cosmetic_df.head()
```

```python
# 각 단계별 unique한 product_id - user_id 조합 집계
price_funnel = cosmetic_df.groupby(['price_range', 'event_type', 'product_id'])['user_id'].nunique()
```

```python
price_funnel = price_funnel.groupby(['price_range', 'event_type']).sum().reset_index()
```

```python
price_funnel
```

```python
# 컬럼명 변경(user_id -> users)
price_funnel = price_funnel.rename(columns={'user_id':'users'})
```

```python
# 'remove_from_cart' 단계 제외
price_funnel = price_funnel.query("event_type != 'remove_from_cart'")
```

```python
# 정렬
price_funnel = price_funnel.groupby('price_range').apply(lambda x: x.sort_values(by='users', ascending=False)).reset_index(drop=True)
```

```python
price_funnel
```

```python
# view를 100%로 두고 전환율 계산
price_funnel['conversion_rate'] = price_funnel['users'] / price_funnel.groupby('price_range')['users'].transform('first')
```

```python
price_funnel
```

```python
# 퍼널 시각화
fig = px.funnel(price_funnel,
                x='conversion_rate',
                y='event_type',
                facet_col='price_range',
                color='price_range',
                category_orders={'price_range':['~3', '3~5', '5~8', '8~']})
fig.update_traces(texttemplate="%{value:.2%}")
```

- 가격대가 높은 상품군에서 구매로까지 이어지는 전환율이 상대적으로 낮게 나타나는 것을 확인할 수 있다.

```python
# 가격대별 이벤트별 유저 수
pivot_temp = price_funnel.pivot(index='price_range', columns='event_type', values='users')
```

```python
pivot_temp
```

```python
# view -> cart 전환율, cart -> purchase 전환율 계산
pivot_temp['view_to_cart'] = pivot_temp['cart'] / pivot_temp['view'] * 100
pivot_temp['cart_to_purchase'] = pivot_temp['purchase'] / pivot_temp['cart'] * 100
```

```python
pivot_temp
```

```python
# index 순서 지정
pivot_temp.reindex(index=['~3', '3~5', '5~8', '8~'])
```

- 가격대가 높아질수록 view -> cart로의 전환율이 떨어지고, cart -> purchase로의 전환율은 3 이하 그룹과 3 초과 그룹 간의 차이는 존재하지만, 3 초과 가격대에서는 cart -> purchase로의 전환율에 큰 차이가 나지 않음을 확인할 수 있다.



# 결론

**분석 결과 요약**
- 상품을 확인(view)한 후 장바구니(cart)에 담는 전환율은 약 12.5%, 상품을 확인(view)한 후 상품이 구매(purchase)까지 이어지는 전환율은 약 4.6%임을 확인할 수 있다.
- 가격대가 높아질수록 상품이 구매(purchase)까지 이어지는 전환율이 감소하는 것을 확인할 수 있다.
- 가격대가 높아질수록 view -> cart로의 전환율이 떨어지고, cart -> purchase로의 전환율은 3 이하 그룹과 3 초과 그룹 간의 차이는 존재하나 3초과 가격대에서는 cart -> purchase로의 전환율에 별다른 차이가 나타나지 않음을 확인할 수 있다.

**개선 전략 제안**
- 가격대가 높은 상품일수록 할인 혜택 등을 강조하여 view -> cart로의 전환율을 1차적으로 높일 수 있도록 신경을 써야 한다.
- 가격대가 3 이하인 상품이라면, cart -> purchase로의 전환율이 다른 가격대와 비교하였을 때 상대적으로 높으므로 카트 내 상품에 대한 알림을 통해 구매를 독려하는 전략을 시도해볼 수 있다.
